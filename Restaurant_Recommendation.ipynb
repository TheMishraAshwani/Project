```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Dataset .csv')

map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=10)

marker_cluster = MarkerCluster().add_to(restaurant_map)
for index, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Restaurant Name']).add_to(marker_cluster)

restaurant_map

df['City'] = df['Address'].str.extract(r'(\w+)$', expand=False)
restaurant_by_City = df.groupby('City')

Votes_rating_by_City = restaurant_by_City['Votes'].mean()
Cuisines_counts_by_City = restaurant_by_City['Cuisines'].value_counts()
Price_range_range_by_City = restaurant_by_city['Price range'].value_counts()

print("Votes:")
print(Votes_rating_by_City)

print("\nTop cuisines by city:")
print(Cuisines_counts_by_City)

print("\n Average Cost for two distribution by city:")
print(Price_range_range_by_City)

plt.figure(figsize=(12, 6))
sns.barplot(x=Votes_rating_by_City.index, y=Votes_rating_by_City.values)
plt.title("Average Votes by City")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.xticks(rotation=90)
plt.show()
```
